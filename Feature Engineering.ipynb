```python
# again we use sk learn train library
from sklearn.model_selection import train_test_split
# for this function we need to provide the data as features and target
Features_train, Features_test,Target_train,Target_test= train_test_split(Features, Target, test_size=0.25)# 25% of data for testing , 75% training
```
