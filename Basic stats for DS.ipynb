```python
plt.rcParams['figure.figsize']=[15,5]
#compute the quartiles
q1= Data.Item_Outlet_Sales.quantile(q=0.25)
q2= Data.Item_Outlet_Sales.quantile(q=0.50)
q3= Data.Item_Outlet_Sales.quantile(q=0.75)

#plot the density 
Data.Item_Outlet_Sales.plot(kind='density')

#to plot the quartile lines
#plt.axvline():plots a line perpendicular to the X-axis at specified location on  x-axis
#x: location on X-axis at which the line is to be plotted
#ymax: specify the height of the line
#colour :colour of the line
#linestyle: specify the style on the line
#label= the name of the line , this name is displayed in the legend

plt.axvline(x=q1, ymax=0.92, color='red', linestyle='--',label='First Quartile')
plt.axvline(x=q2, ymax=0.79, color='blue', linestyle='--',label='Second Quartile')
plt.axvline(x=q3, ymax=0.53, color='green', linestyle='--',label='Third Quartile')

#label the X-axis at the points where we draw the perpendicular lines using plt.xticks()
#locs: specify the location on x-axis
#labels: specify the labels on x-axis
locs,labels =[q1,q2,q3], ['Q1','Q2','Q3'] #get locations and labels
plt.xticks(locs,labels,fontsize=15)
plt.legend()
plt.show()
```

```python
import seaborn as sns
fig,ax= plt.subplots(2,3,figsize=(15,10))
df_cat= Data.select_dtypes(include=object)
#plot a count plot for categorical variables in the data
# use a for loop to plot the count plot for the categoric variable
for variable,subplot in zip(df_cat.columns[1:],ax.flatten()):
  z= sns.countplot(Data[variable],ax=subplot)                     #plot the count plot
  z.set_xticklabels(z.get_xticklabels(),rotation='90')            #rotate the X-axis labels
  z.set_xlabel(variable,fontsize=15)                              #set the size of X- axis labels

#tight_layout() adjusts subplot parameters to give specified padding
#pad specifies the padding (space) b/w the edges and subplots
plt.tight_layout(pad=5.0)
plt.show()
```

```python
def histogram(var):# var will be particular column variance is assigned
  plt.rcParams['figure.figsize']=[15,5]
  var.hist(colour='steelblue', grid=False)
  mu=var.mean()
  mo=var.mode()
  me=var.median()

  plt.axvline(x=mu, color='red', linestyle='--', label='mean')
  plt.axvline(x=me, color='orange', linestyle='--', label='median')
  plt.axvline(x=mo[0], color='black', linestyle='--', label='mode')
  
  plt.legend(prop={'size':20})
```

```python
from matplotlib import gridspec
import seaborn as sns 

plt.rcParams['figure.figsize']=[25,10]
gs = gridspec.GridSpec(2,3, width_ratios=[2.5,2.5,2.5], height_ratios=[0.2,0.01])

a1= plt.subplot(gs[0,0])
a1.set_title('Negatively Skewed Distribution', fontsize=25)
plt.hist(Data['Profit'])
plt.xlabel('Profit',fontsize=15)

b1= plt.subplot(gs[0,1])
b1.set_title('Not Skewed Distribution', fontsize=25)
plt.hist(Data['Item_MRP'])
plt.xlabel('Item_MRP',fontsize=15)

c1= plt.subplot(gs[0,2])
c1.set_title('Positively Skewed Distribution', fontsize=25)
plt.hist(Data['Item_Outlet_Sales'])
plt.xlabel('Item_Outlet_Sales',fontsize=15)


a2 = plt.subplot(gs[1,0])
a2.text(x=0.1,y=0.3,s='Mean < Median < Mode', fontsize=25)
a2.axis('off')


b2 = plt.subplot(gs[1,1])
b2.text(x=0.1,y=0.3,s='Mean = Median = Mode', fontsize=25)
b2.axis('off')


c2 = plt.subplot(gs[1,2])
c2.text(x=0.1,y=0.3,s='Mean > Median > Mode', fontsize=25)
c2.axis('off')

plt.show()
```

```python
import seaborn as sns
plt.rcParams['figure.figsize']=[20,15]

gs = gridspec.GridSpec(3,3, width_ratios=[2.5,2.5,2.5], height_ratios=[0.20,0.01,0.15])

a1=plt.subplot(gs[0,0])
a1.set_title('Negatively Skewed', fontsize=15)
a1= sns.distplot(Data.Profit)

b1=plt.subplot(gs[0,1])
b1.set_title('Not Skewed', fontsize=15)
b1= sns.distplot(Data.Item_MRP)

c1=plt.subplot(gs[0,2])
c1.set_title('Positively Skewed', fontsize=15)
c1= sns.distplot(Data.Item_Outlet_Sales)


a2= plt.subplot(gs[1,0])
a2=sns.boxplot(Data.Profit)
plt.xlim(0,25)


b2= plt.subplot(gs[1,1])
b2=sns.boxplot(Data.Item_MRP)


c2= plt.subplot(gs[1,2])
c2=sns.boxplot(Data.Item_Outlet_Sales)


a3 = plt.subplot(gs[2,0])
a3.text(0.05,0.2, s=str(Data.Profit.describe()),fontsize=15)
a3.axis('off')

b3 = plt.subplot(gs[2,1])
b3.text(0.05,0.2, s=str(Data.Item_MRP.describe()),fontsize=15)
b3.axis('off')

c3 = plt.subplot(gs[2,2])
c3.text(0.05,0.2, s=str(Data.Item_Outlet_Sales.describe()),fontsize=15)
c3.axis('off')
plt.show()
```
