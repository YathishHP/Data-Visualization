# Scale mismatch
*** Standardization , Normalization*** to scale the data 
-as given in belowexample 2 numerical columns has different scales Age Column has 2 digit where as Salary Column has 5 digits this kind of data representation and data scale will leads to one column dominating other in the process of predecting the outcome
- to avoid this kind of bias/influence of one variable over other we apply data standardisation
- standard scaler: here it scales all the variable within a scale such that mean is zero, std1, by applying standard scaler shape of original data is not changed.i,e=xi-mean(x)/standard deviation(x)
- Divide by max value in the column
- min-max normalisation : perforn linear transformation data so that normalised values will be between 0 and 1 i,e to calculate= subract min of column and divide by diff b/w min & max
i,e = xi-min(x)/max(x)-min(x)

```python
import pandas as pd
Data=pd.read_csv('https://raw.githubusercontent.com/Abeyaantrix2Internship/intern/main/Data.csv');Data
```

```python
#binomial: LE
#Nomial: onehot encoding,dummy
#ordinal: Ordinal Encoding

# by using SKLEARN lib or PANDAS lib we can process this kind of data
```

```python
# using SKLEARN library
from sklearn.preprocessing import LabelEncoder
# create the object (instaniate encoder)
LE=LabelEncoder()# here LE object will borrow all properties of Label encoder
Data['Purchased_LE']=LE.fit_transform(Data.Purchased)# this method will apply label encoding and transforms column mentioned, ten particular column will be replaced

print(Data)
```

```python
# a new column will be created
Data['country_LE'] = LE.fit_transform(Data.Country)
```

```python
#ordinal encoding onto country column
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder(categories=[['Spain','France','Germany']])#(order from least to highest)
Data['Country_ord']= OE.fit_transform(Data.Country.values.reshape(-1,1));Data
```

```python
from sklearn.preprocessing import OneHotEncoder
# use column transfer
from sklearn.compose import ColumnTransformer
CT=ColumnTransformer([('encoder',OneHotEncoder(),[0])], remainder='passthrough') # mention column number, remainder passthrough keep other column uneffected drop means delete other column
Data_new= CT.fit_transform(Data)
print(Data_new)
```

```python
from sklearn.preprocessing import OneHotEncoder
# use column transfer
from sklearn.compose import ColumnTransformer
CT=ColumnTransformer([('encoder',OneHotEncoder(),[0,3])], remainder='passthrough') # mention column number, remainder passthrough keep other column uneffected drop means delete other column
Data_new= CT.fit_transform(Data)
print(Data_new)
```
